# Importing required libraries:

```python
import cv2
import numpy as np
from IPython.display import display 
from PIL import Image
```

# Reading Real Image:

```python
image=cv2.imread(r'D:\LGM\disney.jpg')
image
```

```python
image=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
display(Image.fromarray(image))
```

# Displaying Grayscale Image:

```python
gray_img=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
display(Image.fromarray(gray_img))
```

# Displaying Negative Image:

```python
neg_img=cv2.bitwise_not(gray_img)
display(Image.fromarray(neg_img))
```

# Displaying Blurr Image:

```python
blur_img=cv2.GaussianBlur(neg_img,(21,21),0)
display(Image.fromarray(blur_img))
```

# Displaying Inverted Blur Image:

```python
inv_blur=cv2.bitwise_not(blur_img)
display(Image.fromarray(inv_blur))
```

# Finally displaying pencil sketch Image:

```python
sketch=cv2.divide(gray_img,inv_blur,scale=190)
display(Image.fromarray(sketch))
```
